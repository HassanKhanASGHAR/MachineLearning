```python
from scipy.stats import chi2_contingency
# Supongamos que 'target' es tu variable objetivo y 'categorical_var' es una variable categórica
# Crea una tabla de contingencia

for i in col_cat.columns:
    x=col_cat[i]
    print(i)
    contingency_table = pd.crosstab(y, x)
   # print(contingency_table)
  # Realiza la prueba de chi-cuadrado
    chi2, p, dof, expected = chi2_contingency(contingency_table)
    # Imprime los resultados
    print('i=' , i)
    print(f"Chi-cuadrado: {chi2}")
    print(f"Valor p: {p}")
# 3. Aplicar la fórmula de la V de Cramer
    n = contingency_table.sum().sum()
    min_dim = min(contingency_table.shape) - 1
    v_cramer=0
# Manejar el caso donde min_dim es 0
    if min_dim == 0:
       v_cramer = 0 # O np.nan
    else:
       v_cramer = np.sqrt((chi2 / n) / min_dim)

    print(f"V de Cramer para '{i}' vs 'Arrest': {v_cramer:.4f}")
```

```python
#########################################################
################## UNA SIMPLE REGRESIÓN LOGISTICA !!!!!!
#########################################################
# PASO 1. Dividimos train y test al 30 & 70

[X_train, X_test,y_train,y_test] = train_test_split(X,y,test_size=0.3)


#### PASO 2 SELECCION DE VARIABLES (ME LO VOY A SALTAR POR AHORA)

model = ExtraTreesClassifier()
model.fit(X_train, y_train)
model.feature_importances_

Importancia_tree=pd.DataFrame(zip(model.feature_importances_, X_train.columns), columns=['importance', 'feature'])\
    .sort_values('importance', ascending=False).head()

print(Importancia_tree)
## PASO 3 DECIDIMOS EL MODELO (AQUI SIN TUNEAR NI BUSQUEDA PARAMETRICA NI NADA...)

lr = LogisticRegression()
lr.fit(X_train,y_train)

#lr.classes_
#lr.intercept_
#lr.coef_
lr.predict_proba(X_test)
lr.predict(X_test)
lr.score(X_test, y_test)
lr.score(X_train, y_train)
confusion_matrix(y_test, lr.predict(X_test))

confusion_matrix(y_train, lr.predict(X_train))

cm = confusion_matrix(y_test, lr.predict(X_test))

fig, ax = plt.subplots(figsize=(8, 8))
ax.imshow(cm)
ax.grid(False)
ax.xaxis.set(ticks=(0, 1), ticklabels=('Predicted 0s', 'Predicted 1s'))
ax.yaxis.set(ticks=(0, 1), ticklabels=('Actual 0s', 'Actual 1s'))
ax.set_ylim(1.5, -0.5)
for i in range(2):
    for j in range(2):
        ax.text(j, i, cm[i, j], ha='center', va='center', color='red')
plt.show()


print(classification_report(y_test, model.predict(X_test)))

## Paso 4 con validacion cruzada
from sklearn.model_selection import cross_val_score
model=lr
results = cross_val_score(estimator=model, X=X, y=y, cv=5)
print(results)

```

```python
import folium
from folium.plugins import MarkerCluster, HeatMap

# Crear el mapa centrado en la media de coordenadas
center = [df['Latitude'].mean(), df['Longitude'].mean()]
mapa = folium.Map(location=center, zoom_start=11, tiles='CartoDB positron')

# Inicializar clústeres
cluster_arresto = MarkerCluster(name='Arrestos', overlay=True, control=True)
cluster_no_arresto = MarkerCluster(name='No Arrestos', overlay=True, control=True)

# Listas para los heatmaps
coordenadas_arresto = []
coordenadas_no_arresto = []

# Añadir marcadores al mapa
for _, row in df.iterrows():
    coords = [row['Latitude'], row['Longitude']]
    if row['Arrest'] == True:
        folium.Marker(location=coords,
                      popup='Arresto',
                      icon=folium.Icon(color='red', icon='info-sign')).add_to(cluster_arresto)
        coordenadas_arresto.append(coords)
    else:
        folium.Marker(location=coords,
                      popup='No arresto',
                      icon=folium.Icon(color='green', icon='info-sign')).add_to(cluster_no_arresto)
        coordenadas_no_arresto.append(coords)

# Añadir clústeres al mapa
cluster_arresto.add_to(mapa)
cluster_no_arresto.add_to(mapa)

# Añadir capas de HeatMap
HeatMap(coordenadas_arresto, name='Heatmap Arrestos', radius=10, blur=15).add_to(mapa)
HeatMap(coordenadas_no_arresto, name='Heatmap No Arrestos', radius=10, blur=15, gradient={0.4: 'lime', 0.65: 'yellow', 1: 'orange'}).add_to(mapa)

# Añadir control de capas
folium.LayerControl().add_to(mapa)

# Guardar el mapa final
mapa.save('mapa_arrestos_y_no_arrestos.html')
```
