# Proceso completo ML
- **Data preparation** (80% tiempo)
  1. Exploring
  2. Cleansing
  3. Transformation
- **Model training** (20% tiempo)
  1. Feature selection: _¿qué variables explicativas voy a usar y por qué?_
  2. Model validation: _¿qué estructura de validación voy a usar?_
  3. Metric analysis: _¿en base a qué métrica quiero analizar mis resultados? ¿me interesa ganar en especificidad? ¿me afectan mucho los falsos negativos?_
  4. Model selection: _¿qué técnica predictiva voy a aplicar?_
  5. Model tuning: _ajuste de hiperparámetros para la técnica escogida_

## Ejemplo 1
Cargar el conjunto de datos _wine_, conocer de qué tratan las variables, aplicar una regresión logística usando una estructura tr/ts con tr=75%, y validar el modelo en base al accuracy

```python
# Importar las librerías necesarias
from sklearn.datasets import load_wine
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
# Cargar el conjunto de datos 'wine'
data = load_wine() #conjunto de datos con variable objetivo multi-clase, para evaluar la calidad del vino: Clase 0, Clase 1, Clase 2
X = data.data  # variables explicativas
y = data.target  # variable objetivo

# Conocer las variables
print("Variables explicativas: ", data.feature_names)
print("Variable objetivo: ", data.target_names) 

# Dividir el conjunto de datos en entrenamiento y prueba (train/test)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)

# Inicializar el modelo de regresión logística
model = LogisticRegression()

# Entrenar el modelo
model.fit(X_train, y_train)

# Realizar predicciones sobre el conjunto de prueba
y_pred = model.predict(X_test)
y_pred_prob= model.predict_proba(X_test)
# Calcular el accuracy del modelo
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy del modelo: {accuracy:.4f}")
```

## Ejemplo 2
Aplicar un proceso similar al del Ejemplo 1, pero en este caso, trabajar sobre datos normalizados

```python
# Importar las librerías necesarias
from sklearn.datasets import load_wine
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import StandardScaler

# Cargar el conjunto de datos 'wine'
data = load_wine()
X = data.data  # variables explicativas
y = data.target  # variable ovjetivo

# Conocer las variables
print("Variables explicativas: ", data.feature_names)
print("Variable objetivo: ", data.target_names) 

# Normalizar las características (escala entre 0 y 1 o media 0, desviación 1)
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Dividir el conjunto de datos en entrenamiento y prueba (train/test)
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.25, random_state=42)

# Inicializar el modelo de regresión logística
model = LogisticRegression(max_iter=1000)

# Entrenar el modelo
model.fit(X_train, y_train)

# Realizar predicciones sobre el conjunto de prueba
y_pred = model.predict(X_test)

# Calcular el accuracy del modelo
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy del modelo: {accuracy:.4f}")
```

## OJO!!
¿Encuentras alguna diferencia en el proceso de transformación entre los Ejemplos 1 y 2?
Las transformaciones de datos deben hacerse después de dividir los datos en train y test. 
1. Evitar la contaminación de datos (_data leakage_): el escalador tiene acceso a información de los datos de prueba durante el ajuste, lo que nos puede llevar a modelos no generalizables, ya que los datos de prueba influyen indirectamente en el entrenamiento.
2. Simular un _escenario real_, en el que los datos de prueba nunca estarán disponibles durante el entrenamiento

# Ejemplo 3
Trabajar con un problema de predicción de variable continua, *california_housing*, sobre información de viviendas de California y su precio medio. Trabaja con datos estandarizados y evalúa el modelo de regresión usando el MSE. El conjunto de datos lo tienes disponible en formato _.csv_.

```python
# Importar las librerías necesarias
import pandas as pd
from sklearn.datasets import fetch_california_housing
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

# 1. Cargar el conjunto de datos California Housing desde un csv
df = pd.read_csv('/Users/biblioadmin/Library/CloudStorage/OneDrive-UniversidadComplutensedeMadrid(UCM)/Documentos/Titulo Propio/Machine Learning/Redes Neuronales/Python/Datasets/california_housing.csv')

# Mostrar las primeras filas para ver cómo se ve el conjunto de datos
print(df.head())

# 2. Separar las características (X) y la variable objetivo (y)
# 'MedHouseVal' es la columna con el precio medio de la vivienda (variable objetivo)
X = df.drop('MedHouseVal', axis=1)  # Las características (todas las columnas excepto 'MedHouseVal')
y = df['MedHouseVal']  # La variable objetivo

# 3. Dividir el conjunto de datos en entrenamiento (75%) y prueba (25%)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)

# 4. Estandarizar los datos
# Se crea un objeto de StandardScaler para escalar las características (normalizar los valores)
scaler = StandardScaler()

# Ajustamos el scaler a los datos de entrenamiento y transformamos los datos de entrenamiento y prueba
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# 5. Entrenar el modelo de regresión lineal
# Creamos una instancia del modelo de regresión lineal
model = LinearRegression()

# Entrenamos el modelo con los datos estandarizados
model.fit(X_train_scaled, y_train)

# 6. Realizar predicciones en el conjunto de prueba
y_pred = model.predict(X_test_scaled)

# 7. Evaluar el modelo utilizando el Error Cuadrático Medio (MSE)
# Calculamos el MSE entre las predicciones y los valores reales
mse = mean_squared_error(y_test, y_pred)

# Mostrar el resultado del MSE
print(f"Error Cuadrático Medio (MSE): {mse:.4f}")
```

# Ejemplo 4
Hacer un análisis predictivo del conjunto de datos _diamonds_ disponible en _sns_. Trabajar la regresión con datos estandarizados, sólo con variables numéricas y evaluar los resultados con R2. Usar una estructura de validación tr/ts dejando el 20% de los datos para la prueba.

```python
# Importar las librerías necesarias
import pandas as pd
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score

# 1. Cargar el conjunto de datos diamonds desde seaborn y convertirlo a un DataFrame
df = sns.load_dataset('diamonds')

# Mostrar las primeras filas para ver cómo se ve el conjunto de datos
print(df.head())

# 2. Preparar los datos: eliminar columnas no numéricas('cut', 'color', 'clarity')
df = df.drop(['cut', 'color', 'clarity'], axis=1)

# Separar las variables explicativas (X) y la variable objetivo (y)
X = df.drop('price', axis=1)  # (todas las columnas de DF excepto 'price')
y = df['price']              
# 3. Dividir el conjunto de datos en entrenamiento (80%) y prueba (2%)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=SEED)

# 4. Estandarizar los datos: crear el objeto que usaremos para estandarizar
scaler = StandardScaler()

# Ajustar el escalador con los datos de entrenamiento y transformar los datos
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# 5. Entrenar el modelo de regresión lineal
# Escoger el modelo de regresion lineal: la variable objetico es continua
model = LinearRegression()

# Entrenar el modelo con los datos estandarizados
model.fit(X_train_scaled, y_train)

# 6. Realizar predicciones en el conjunto de prueba
y_pred = model.predict(X_test_scaled)

# 7. Evaluar el modelo utilizando el coeficiente de determinación (R²)
r2 = r2_score(y_test, y_pred)

# Mostrar el resultado de R²
print(f"Coeficiente de Determinación (R²): {r2:.4f}")
```
